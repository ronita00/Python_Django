```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters for the hyperbola
a = 5  # Controls horizontal stretch
b = 3  # Controls vertical stretch

# Generate x values
x = np.linspace(-10, 10, 500)

# Calculate y values for the hyperbola
y_upper = np.sqrt((x**2 / a**2 - 1) * b**2)
y_lower = -y_upper

# Plot the hyperbola
plt.plot(x, y_upper, color='blue')
plt.plot(x, y_lower, color='red')

# Add axes and labels
plt.axhline(0, color='black', linewidth=0.5)  # X-axis
plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
plt.title("Hyperbola: x^2/a^2 - y^2/b^2 = 1")
plt.grid(True)

# Show the plot
plt.show()
```
