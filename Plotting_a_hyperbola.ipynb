```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the hyperbola: a^2 and b^2
a = 5  # Horizontal stretch
b = 3  # Vertical stretch

# Generate x values
x = np.linspace(-10, 10, 500)

# Calculate y values for the upper and lower branches of the hyperbola
y_upper = np.sqrt((x**2 / a**2 - 1) * b**2)
y_lower = -y_upper

# Create the plot
plt.figure(figsize=(8, 8))
plt.plot(x, y_upper, label='Upper branch', color='blue')
plt.plot(x, y_lower, label='Lower branch', color='red')
plt.axhline(0, color='black', linewidth=0.5)  # X-axis
plt.axvline(0, color='black', linewidth=0.5)  # Y-axis
plt.grid(True, linestyle='--', alpha=0.6)

# Add labels and title
plt.title(r"Hyperbola: $\frac{x^2}{a^2} - \frac{y^2}{b^2} = 1$", fontsize=16)
plt.xlabel("X-axis", fontsize=12)
plt.ylabel("Y-axis", fontsize=12)

# Set aspect ratio for equal scaling
plt.gca().set_aspect('equal', adjustable='box')

# Show the plot
plt.legend()
plt.show()
```
