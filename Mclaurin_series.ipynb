```python
import math

def e_x(x, terms=10):
    approximation = 0
    for i in range(terms):
        approximation += (x ** i) / math.factorial(i)
    return approximation

x = int(input("Enter x: "))
for n in range(1, 100):
    approx_value = e_x(x, n)
    exact_value = math.exp(x)
    truncation_error = exact_value - approx_value

    print(f"e^{x} (for {n} terms) = {approx_value}")
    print(f"Truncation Error = {truncation_error} for {n} terms\n")
```

```python
import numpy as np
import pandas as pd
import math

# Vectorized factorial for efficiency
fac = np.vectorize(factorial)

def e_x(x, terms):
    """Approximates e^x using a given number of terms of
    the Maclaurin series.
    """
    n = np.arange(terms)
    return np.sum((x ** n) / fac(n))

# Read input values from CSV
input_df = pd.read_csv('https://raw.githubusercontent.com/ronita00/Python_Django/refs/heads/main/input_df.csv')

# Initialize an empty list to store the results
results = []

# Iterate over each row in the input DataFrame
for index, row in input_df.iterrows():
    x = row['x']
    terms = int(row['terms'])

    # Calculate Maclaurin series sum and error
    maclaurin_sum = e_x(x, terms)
    actual_value = math.exp(x)
    error = actual_value - maclaurin_sum

    # Append the results to the list
    results.append({
        'x': x,
        'terms': terms,
        'Maclaurin Sum': maclaurin_sum,
        'Actual Value': actual_value,
        'Error': error
    })

# Create a DataFrame from the results
output_df = pd.DataFrame(results)

# Write output to CSV
output_df.to_csv('output.csv', index=False)

print("Calculation complete. Results saved to output.csv.")


```
